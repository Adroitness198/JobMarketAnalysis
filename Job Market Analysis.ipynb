```python
import pandas as pd

df = pd.read_csv(r"C:\Users\sinet\Desktop\Job Market Analysis\glassdoor.csv")
print(df.shape)
df.head()
```

```python
print(df.columns.tolist())
```

```python
columns_we_need = [
    'gaTrackerData.jobTitle',
    'gaTrackerData.location',
    'gaTrackerData.industry',
    'gaTrackerData.sector',
    'gaTrackerData.empSize',
    'header.employerName',
    'header.rating',
    'header.payLow',
    'header.payHigh',
    'header.payMed',
    'header.payPeriod',
    'overview.size',
    'overview.revenue',
    'overview.type',
    'map.country',
]

df_clean = df[columns_we_need]
print(df_clean.shape)
df_clean.head()
```

```python
df_clean.info()
```

```python
df_clean.isnull().sum()
```

```python

df_clean = df_clean.drop(columns=[
    'header.payLow',
    'header.payHigh', 
    'header.payMed',
    'header.payPeriod'
])

df_clean = df_clean.dropna(subset=['gaTrackerData.jobTitle'])

print(df_clean.shape)
df_clean.isnull().sum()
```

```python
df_clean = df_clean.rename(columns={
    'gaTrackerData.jobTitle': 'job_title',
    'gaTrackerData.location': 'location',
    'gaTrackerData.industry': 'industry',
    'gaTrackerData.sector': 'sector',
    'gaTrackerData.empSize': 'company_size',
    'header.employerName': 'company_name',
    'header.rating': 'company_rating',
    'overview.size': 'size',
    'overview.revenue': 'revenue',
    'overview.type': 'company_type',
    'map.country': 'country'
})

df_clean.head()
```

```python
top_jobs = df_clean['job_title'].value_counts().head(20)
print(top_jobs)
```

```python
top_industries = df_clean['industry'].value_counts().head(10)
print(top_industries)
```

```python
top_countries = df_clean['country'].value_counts().head(10)
print(top_countries)
```

```python
print(df_clean['company_rating'].describe())
```

```python
top_sizes = df_clean['company_size'].value_counts()
print(top_sizes)
```

```python
df_clean.to_csv(r"C:\Users\sinet\Desktop\Job Market Analysis\glassdoor_cleaned.csv", index=False)
print("Saved!")
```

```python
import matplotlib.pyplot as plt

top_jobs = df_clean['job_title'].value_counts().head(15)

plt.figure(figsize=(12, 6))
top_jobs.plot(kind='barh', color='steelblue')
plt.title('Top 15 Most In-Demand Jobs')
plt.xlabel('Number of Listings')
plt.ylabel('Job Title')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig(r"C:\Users\sinet\Desktop\Job Market Analysis\top_jobs.png")
plt.show()
```

```python
top_industries = df_clean['industry'].value_counts().head(15)

plt.figure(figsize=(12, 6))
top_industries.plot(kind='barh', color='coral')
plt.title('Top 15 Industries Hiring')
plt.xlabel('Number of Listings')
plt.ylabel('Industry')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig(r"C:\Users\sinet\Desktop\Job Market Analysis\top_industries.png")
plt.show()
```

```python

za_jobs = df_clean[df_clean['country'] == 'ZA']

print(f"Total SA job listings: {len(za_jobs)}")

top_za_jobs = za_jobs['job_title'].value_counts().head(15)

plt.figure(figsize=(12, 6))
top_za_jobs.plot(kind='barh', color='green')
plt.title('Top 15 Most In-Demand Jobs in South Africa')
plt.xlabel('Number of Listings')
plt.ylabel('Job Title')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig(r"C:\Users\sinet\Desktop\Job Market Analysis\top_za_jobs.png")
plt.show()
```

```python
top_za_industries = za_jobs['industry'].value_counts().head(15)

plt.figure(figsize=(12, 6))
top_za_industries.plot(kind='barh', color='goldenrod')
plt.title('Top 15 Industries Hiring in South Africa')
plt.xlabel('Number of Listings')
plt.ylabel('Industry')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig(r"C:\Users\sinet\Desktop\Job Market Analysis\top_za_industries.png")
plt.show()
```

```python

df_valid = df_clean[df_clean['company_rating'] > 0]
za_valid = df_valid[df_valid['country'] == 'ZA']
global_valid = df_valid[df_valid['country'] != 'ZA']

print(f"SA Average Rating: {za_valid['company_rating'].mean():.2f}")
print(f"Global Average Rating: {global_valid['company_rating'].mean():.2f}")

plt.figure(figsize=(10, 5))
plt.hist(global_valid['company_rating'], bins=20, alpha=0.5, color='steelblue', label='Global')
plt.hist(za_valid['company_rating'], bins=20, alpha=0.7, color='green', label='South Africa')
plt.title('Company Ratings: South Africa vs Global')
plt.xlabel('Rating')
plt.ylabel('Number of Companies')
plt.legend()
plt.tight_layout()
plt.savefig(r"C:\Users\sinet\Desktop\Job Market Analysis\ratings_comparison.png")
plt.show()
```

```python
za_sizes = za_jobs['company_size'].value_counts()

plt.figure(figsize=(10, 5))
za_sizes.plot(kind='bar', color='steelblue')
plt.title('Company Sizes Hiring in South Africa')
plt.xlabel('Company Size')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(r"C:\Users\sinet\Desktop\Job Market Analysis\za_company_sizes.png")
plt.show()
```

```python

za_jobs_clean = za_jobs[~za_jobs['company_size'].isin(['-1-0', '10000--1'])]

size_order = ['1-50', '51-200', '201-500', '501-1000', '1001-5000', '5001-10000']

za_sizes_clean = za_jobs_clean['company_size'].value_counts().reindex(size_order)

plt.figure(figsize=(10, 5))
za_sizes_clean.plot(kind='bar', color='steelblue')
plt.title('Company Sizes Hiring in South Africa')
plt.xlabel('Company Size (Employees)')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(r"C:\Users\sinet\Desktop\Job Market Analysis\za_company_sizes.png")
plt.show()
```
